Для обмена информацией с SOAP API, в языке Python3, удобно использовать модуль zeep, избавляющий от необходимости формировать и парсить xml собственноручно. 
В данной документации, для наглядности, используется модуль requests, с помощью zeep только формируется xml запрос.

```python
import zeep
import requests
from lxml.etree import tostring
from bs4 import BeautifulSoup
```

```python
client = zeep.Client('http://localhost:5000/?wsdl')
```

```python
def send_xml(nd):
    xml=tostring(nd, pretty_print=True).decode()
    print('IN:\n',xml)
    res=requests.post('http://localhost:5000',data=xml, headers={'Content-Type': 'application/xml'})
    print('HTTP:\n',res.status_code, res.reason)
    print('OUT:\n',BeautifulSoup(res.content, "xml").prettify())
```

```python
send_xml(client.create_message(client.service, 'get_movies'))
```

```python
send_xml(client.create_message(client.service, 'get_movie', movie_id='0000000'))
```

```python
send_xml(client.create_message(client.service, 'get_movie_by', year=2010))
```

```python
send_xml(client.create_message(client.service, 'get_movie_by', year=2010, genre='Documentary'))
```

```python
send_xml(client.create_message(client.service, 'add_movie', movie_id='0000000', title='test'))
```

```python
send_xml(client.create_message(client.service, 'update_movie', movie_id='0000000', title='TEST'))
```

```python
send_xml(client.create_message(client.service, 'delete_movie', movie_id='0000000'))
```
